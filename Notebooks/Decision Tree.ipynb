# Decision Tree Classifier

##### A decision tree is a flowchart-like structure in which each internal node represents a "test" on an attribute (e.g. whether a coin flip comes up heads or tails), each branch represents the outcome of the test, and each leaf node represents a class label (decision taken after computing all attributes).

- It is a tree-structured classifier.

- There are two nodes, which are the Decision Node and Leaf Node.

- A decision tree simply asks a question, and based on the answer (Yes/No), it further split the tree into subtrees.

- Root node is selected bt ASM (Attribute selection measure - (Information gain, Gini index))

```python
# importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
# Reading the dataset (csv file)
df = pd.read_csv("loan_prediction.csv")
```

```python
# Visualizing 1st 5 rows
df.head()
```

```python
# Finding shape
df.shape
```

```python
# Descriptive stats
df.describe(include='all')
```

```python
# Finding for null values
df.isnull().sum()
```

```python
# Removing unwanted columns
df.drop(columns=['Loan_ID','Gender','Dependents','Self_Employed'],inplace=True)
```

```python
# Handling null values
df['Credit_History'].fillna(df['Credit_History'].mean(),inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean(),inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].mean(),inplace=True)
df['Married'].fillna("Yes",inplace=True)
```

```python
df.isnull().sum()
```

```python
# Checking data types
df.info()
```

```python
# Encoding
le = LabelEncoder()

df['Married'] = le.fit_transform(df['Married'])
df['Education'] = le.fit_transform(df['Education'])
df['Property_Area'] = le.fit_transform(df['Property_Area'])
```

```python
df.head()
```

```python
# Spliting dependent and independent variable
x = df.drop('Loan_Status',axis=1)
y = df['Loan_Status']
```

```python
# Spliting data for train and test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
# Initializing DT
dt = DecisionTreeClassifier()
```

```python
# Training model
dt.fit(x_train,y_train)
```

```python
# Testing model
pred = dt.predict(x_test)
```

```python
# Evaluating model performance
print(metrics.classification_report(y_test,pred))
```

```python
# Evaluating model performance
metrics.confusion_matrix(y_test,pred)
```
