# Logistic Regression

- Logistic regression is one of the most popular Machine Learning algorithms (Supervised Learning). It is used for 
predicting the categorical dependent variable using a given set of independent variables.

- The outcome must be a categorical or discrete value. It can be either Yes or No, 0 or 1, true or False, etc. but instead 
of giving the exact value as 0 and 1, it gives the probabilistic values which lie between 0 and 1.

- "S" shaped curve is used, which predicts two maximum values (0 or 1).

Logistic Function (Sigmoid Function):

- The sigmoid function is a mathematical function used to map the predicted values to probabilities (maps any real value 
into another value within a range of 0 and 1).

```python
# Reading the dataset (csv file)
df = pd.read_csv("Datasets/Social_Network_Ads.csv")
```

```python
# Visualizing 1st 5 rows
df.head()
```

```python
# Descriptive stats
df.describe()
```

```python
# Finding shape
df.shape
```

```python
# Finding for null values
df.isnull().sum()
```

```python
# Droping unwanted column
df.drop(columns=['User ID'],inplace=True)
```

```python
# Feature mapping is done for encoding
df['Gender'].replace({"Male":1,"Female":0},inplace=True)
```

```python
# Spliting dependent and independent variable
x = df.drop('Purchased',axis=1)
y = df['Purchased']
```

```python
# Spliting data for train and test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
# Initializing logistic regression
log_reg = LogisticRegression()
```

```python
# Training model
log_reg.fit(x_train,y_train)
```

```python
# Testing model
pred = log_reg.predict(x_test)
```

```python
# Evaluating model performance
print(metrics.classification_report(y_test,pred))
```

```python
# Evaluating model performance
metrics.confusion_matrix(y_test,pred)
```
