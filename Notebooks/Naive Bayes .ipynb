```python
# importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import model_selection
from sklearn import naive_bayes
from sklearn import metrics
```

```python
# Read the data
data = pd.read_csv(r"house-votes-84.csv")
```

```python
# Visualizing 1st 5 rows
data.head()
```

```python
# Finding shape
data.shape
```

```python
# # Finding for null values
data.isnull().sum()
```

```python
# Checking data types
data.info()
```

```python
# Spliting dependent and independent variable
x = data.iloc[:,1:]
y = data.iloc[:,0]
```

```python
names = x.columns
```

```python
names
```

```python
# Manual Encoding
x = np.where(x == 'y',1,x)
```

```python
# Manual Encoding
x = np.where(x == 'n',0,x)
```

```python
# Manual Encoding
x = np.where(x == "?",1,x)
```

```python
# Creating DataFrame
x = pd.DataFrame(x,columns=names)
```

```python
x.head()
```

```python
# Discriptive stats
x.describe()
```

```python
# Data spliting into test and train
x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Initializing Naive bayes model
nb = naive_bayes.GaussianNB()
```

```python
# fitting the model
nb.fit(x_train,y_train)  
```

```python
# Testing 
pred = nb.predict(x_test) 
```

```python
# Evaluating model performance
metrics.confusion_matrix(pred,y_test)
```

```python
# Evaluating model performance
print(metrics.classification_report(pred,y_test)) 
```
