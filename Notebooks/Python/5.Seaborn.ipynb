### Distribution plot

```python
# Distribution plot shows us the frequency
```

```python
data.head()
```

```python
sns.displot(data['total_bill'])
```

```python
sns.displot(data.total_bill,kind="kde")
```

```python
sns.scatterplot(data.total_bill,data.tip)
```

```python
sns.jointplot(data.total_bill,data.tip)
```

```python
sns.jointplot(data.total_bill,data.tip,kind="kde")
```

```python
sns.boxplot(data.total_bill)
```

```python
sns.barplot(data.total_bill,data.tip)
```

```python
sns.barplot(y = data.total_bill,x = data.sex)
```

###Heatmap

```python
# By using the hetmap we find out the corelation
```

```python
a = data.corr()
```

```python
sns.heatmap(a)
```

```python
sns.heatmap(a,annot=True)
```

```python
sns.pairplot(data)
```

### Line Plot

```python
sns.relplot(data = data, x = "total_bill", y = "tip", kind = "line")
```
