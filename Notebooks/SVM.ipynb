# SVM 

#### SVM objective is to define a hyperplane which divides the data into classes, hyperplane in 2D is called a line, in 3D called a plane 

hyperplane should be at equal distance from both the classes
 
hyperplane will be a linear or non linear plane based on seperation of classes in the dataset

for linear we use linear plane,for non- linear we use kernal trick


- The goal of the SVM algorithm is to create the best line or decision boundary that can segregate n-dimensional space 
into classes so that we can easily put the new data point in the correct category in the future. 

- This best decision boundary is called a hyperplane.

- SVM chooses the extreme points/vectors that help in creating the hyperplane. 

- These extreme cases are called as support vectors

```python
import pandas as pd
# importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
plt.imshow(plt.imread(r'download (4).png'))
```

```python
# Reading the dataset (csv file)
data = pd.read_csv(r'Datasets/Iris.csv')
```

```python
# Visualizing 1st 5 rows
data.head()
```

```python
# Descriptive stats
data.describe()
```

```python
# Finding for null values
data.isnull().sum()
```

```python
# Removing unwanted columns
data.drop(columns=['Id'],inplace=True)
```

```python
# Spliting dependent and independent variable
x = data.iloc[:,:-1]
y = data['Species']
```

```python
# Spliting data for train and test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

# SVM - Linear kernals

```python
# SVM is a model from sklearn library
from sklearn import svm      
```

```python
# Initializing SVM
model = svm.SVC(kernel="linear") #if data is linearly sepearable 
```

```python
# Training model
MSVC = model.fit(x_train,y_train)
```

```python
# Testing model
pred = MSVC.predict(x_test)
```

```python
# Evaluating model performance
metrics.confusion_matrix(pred,y_test)
```

```python
# Evaluating model performance
metrics.accuracy_score(pred,y_test)
```

# SVM Non-Linear Kernals

```python
# when we apply non-linear models on linearly seperable data,model will be specfic in prediction
# high variance and overfitted model will be generated
# non-linear methods will be applied only on the data,which is non-linearly seperable
```

```python
# Initializing SVM
model = svm.SVC(kernel="poly")   # polynomial kernal = usually overfitted model,prediction is specific in nature
```

```python
# Training model
fit = model.fit(x_train,y_train)
```

```python
# Testing model
pred = fit.predict(x_test)
```

```python
metrics.confusion_matrix(pred,y_test)
```

```python
metrics.accuracy_score(pred,y_test)
```
